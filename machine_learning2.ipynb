```python
import numpy as np
import matplotlib.pyplot as plt

# Define the ODE function
def dydx(t, y):
    return -2 * y

# Initial condition
y0 = 1

# Time settings
t0 = 0
tf = 5
dt = 0.01
t_points = np.arange(t0, tf + dt, dt)

# Euler's method
y_points = []
y = y0
for t in t_points:
    y_points.append(y)
    y = y + dt * dydx(t, y)

# Plot the solution
plt.plot(t_points, y_points, label="Euler's Method")
plt.xlabel('t')
plt.ylabel('y')
plt.title("Solution of dy/dx = -2y using Euler's Method")
plt.legend()
plt.grid()
plt.show()
```

```python
import numpy as np
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt

# Define the ODE function
def system(t, y):
    y1, y2 = y
    dydt = [y2, -3*y2 - 2*y1]
    return dydt

# Initial conditions
y0 = [1, 0]  # y1(0) = 1, y2(0) = 0

# Time span
t_span = (0, 10)

# Solve the ODE
solution = solve_ivp(system, t_span, y0, t_eval=np.linspace(0, 10, 100))

# Plot the solution
plt.plot(solution.t, solution.y[0], label='y1(t)')
plt.plot(solution.t, solution.y[1], label='y2(t)')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Solution of Second-Order ODE')
plt.legend()
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the system of ODEs
def dydx(x, y):
    y1, y2 = y
    dy1dx = y2
    dy2dx = -3 * y2 - 2 * y1
    return np.array([dy1dx, dy2dx])

# Initial conditions
x0 = 0
y0 = [1, 0]  # y1(0) = 1, y2(0) = 0

# Time settings
h = 0.1  # step size
x_end = 10  # end point

# Create an array of x values
x_values = np.arange(x0, x_end + h, h)

# Initialize y values
y_values = np.zeros((len(x_values), 2))#[]
y_values[0] = y0

# Euler's method
for i in range(1, len(x_values)):
    y_values[i] = y_values[i-1] + h * dydx(x_values[i-1], y_values[i-1])

# Plot the solutions
plt.plot(x_values, y_values[:, 0], label='y1(t)')
plt.plot(x_values, y_values[:, 1], label='y2(t)')
plt.xlabel('x')
plt.ylabel('y')
plt.title("Solution of Second-Order ODE using Euler's Method")
plt.legend()
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import sciann as sn
from sciann.utils.math import diff

# Define the variables
x = sn.Variable('x')
y = sn.Functional('y', x, 8*[20], 'tanh')  # 8 layers of 20 neurons each with tanh activation

# Define the first and second derivatives
y_x = diff(y, x)
y_xx = diff(y_x, x)

# Define the ODE
L1 = y_xx + 3*y_x + 2*y
#define bc
bc = y-1
bc2 = y_x
# Define the data for training
x_data = np.linspace(0, 10, 1000)

# Define the model
model = sn.SciModel(x, [L1,bc,bc2])

# Compile the model with the initial condition


# Train the model
model.train(x_data,3*['zero'], epochs=100)

# Make predictions
y_pred = y.eval(model, [x_data])

# Plot the solution
plt.plot(x_data, y_pred, label='Predicted y(t)')
plt.xlabel('x')
plt.ylabel('y')
plt.title("Solution of Second-Order ODE using SciANN")
plt.legend()
plt.grid()
plt.show()
```

```python
import numpy as np
import qutip

def qubit_integrate(w, theta, gamma1, gamma2, psi0, tlist):
    # operators and the hamiltonian
    sx = qutip.sigmax()
    sy = qutip.sigmay()
    sz = qutip.sigmaz()
    sm = qutip.sigmam()
    H = w * (np.cos(theta) * sz + np.sin(theta) * sx)
    # collapse operators
    c_op_list = []
    n_th = 0.5 # temperature
    rate = gamma1 * (n_th + 1)
    if rate > 0.0: c_op_list.append(np.sqrt(rate) * sm)
    rate = gamma1 * n_th
    if rate > 0.0: c_op_list.append(np.sqrt(rate) * sm.dag())
    rate = gamma2
    if rate > 0.0: c_op_list.append(np.sqrt(rate) * sz)
    # evolve and calculate expectation values
    output = qutip.mesolve(H, psi0, tlist, c_op_list, [sx, sy, sz])
    return output.expect[0], output.expect[1], output.expect[2]

## calculate the dynamics
w     = 1.0 * 2 * np.pi  # qubit angular frequency
theta = 0.2 * np.pi      # qubit angle from sigma_z axis (toward sigma_x axis)
gamma1 = 0.5             # qubit relaxation rate
gamma2 = 0.2             # qubit dephasing rate
# initial state
a = 1.0
psi0 = (a*qutip.basis(2, 0) + (1-a)*qutip.basis(2, 1))/np.sqrt(a**2 + (1-a)**2)
tlist = np.linspace(0, 4, 250)
#expectation values for ploting
sx, sy, sz = qubit_integrate(w, theta, gamma1, gamma2, psi0, tlist)
```

```python
from qutip import *   
  
from mpl_toolkits.mplot3d import Axes3D   
from matplotlib import cm   
from pylab import *  

#setup constants: 
N = 20 # size of the Hilbert space 
alpha=linspace(0.01,3,60) # values for the coherent state amplitude 
a = destroy(N) #annihilation operator 
D = displace(N,alpha) # Displacement 
xvec = linspace(-9,9,300) 
X,Y = meshgrid(xvec, xvec)

k=1 # counter 
for x in alpha: 

    # Define the cat state. It is a coherent superposition of two coherent states. 
    # Each coherent state is created by the displacement operator acting on the vacuum. 
    # the unit() function makes sure that the overall state is normalized. 

    psi=(displace(N,x)*basis(N,0)-displace(N,-1*x)*basis(N,0)).unit();

    # plot the wigner function 
    W=wigner(psi,xvec,xvec) 
    fig =figure() 
    ax = Axes3D(fig,azim=-62.5,elev=25) 
    ax.plot_surface(X, Y, W, rstride=2, cstride=2, cmap=cm.jet,lw=.1) 
    ax.set_xlim3d(-9,9) 
    ax.set_xlim3d(-9,9) 
    ax.set_zlim3d(-.2,0.2) 
    ax.set_axis_off() 
    ax.set_frame_on('false') 
    title(r'$| \psi >= \frac{1}{\sqrt{2}}(|\alpha>-|-\alpha>)$'+r' $\alpha=$'+str(round(x,2))) 
    
```

```python
omega = 1.0  # Angular frequency

# Create the Hamiltonian for a qubit in a magnetic field along X-axis
H = 0.5 * omega * sigmax()

# Initial state |0⟩
psi_0 = basis(2, 0)

# Time points for simulation
times = linspace(0, 10, 100)

# Perform unitary evolution
result = mesolve(H, psi_0, times, [], [])

# Calculate expectation values of Pauli operators
expect_x = expect(sigmax(), result.states)
expect_y = expect(sigmay(), result.states)
expect_z = expect(sigmaz(), result.states)

# Plot the results
plot(times, expect_x, label='<σx>')
plot(times, expect_y, label='<σy>')
plot(times, expect_z, label='<σz>')
xlabel('Time')
ylabel('Expectation Value')
legend()
title('Expectation Values of Pauli Operators')
show()
```
